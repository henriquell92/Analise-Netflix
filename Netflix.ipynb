```python
#Import das Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#Fazendo a leitura do DataSet
df = pd.read_csv('netflix_pric_countries.csv', sep=',')
df
```

```python
#Verificando se há valores nullos
df.isnull().sum()
```

```python
#Filtrando apenas as colunas de interesse para a analise.
Netflix = df[['Country','Cost Per Month - Basic ($)','Cost Per Month - Standard ($)','Cost Per Month - Premium ($)']]
Netflix
```

```python
#Colocando a coluna em ordem decrescente
NetflixDecrs = Netflix.sort_values('Cost Per Month - Premium ($)', ascending = False)
NetflixDecrs
```

```python
#Selecionando os 10 paises com Assinatura Mais cara
NetflixTop10 = NetflixDecrs.head(10)
NetflixTop10
```

```python
#Criando um gráfico para analisar os top10 valores mais altos por país.
plt.barh(NetflixTop10['Country'],NetflixTop10['Cost Per Month - Premium ($)'], label='info')
plt.xlabel('Cost Per Month - Premium ($)')
plt.ylabel('Country')
plt.title('Country x Premium')
plt.legend(['Valores'])
```
